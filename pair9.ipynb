2.1

Intervalo de Confianza para la Esperanza de Vida

Contexto: Trabajas en una organización de salud global y deseas estimar el intervalo de confianza al 95% para la esperanza de vida promedio de cada continente.

Nota: Cuidado porque puede que os salgan algunos nulos, genera un código para que no salgan nulos en los intervalos de confianza.

```python
media_lifeexpectancy_i = df["lifeexpectancy_i"].mean()

error_lifeexpectancy_i = stats.sem(df["lifeexpectancy_i"])

nivel_confianza_lifeexpectancy_i = 0.95

grados_libertad_lifeexpectancy_i = len(df["lifeexpectancy_i"]) - 1

valor_critico = stats.t.ppf((1 + nivel_confianza_lifeexpectancy_i) / 2, df=grados_libertad_lifeexpectancy_i)

limite_inferior_lifeexpectancy_i = media_lifeexpectancy_i - valor_critico * error_lifeexpectancy_i
limite_superior_lifeexpectancy_i = media_lifeexpectancy_i + valor_critico * error_lifeexpectancy_i

print("Intervalo de Confianza para la esperanza de vida:")
print(f"Media Muestral: {np.round(media_lifeexpectancy_i, 2)}")
print(f"Error Estándar: {np.round(error_lifeexpectancy_i, 2)}")
print(f"Nivel de Confianza: {nivel_confianza_lifeexpectancy_i}")
print(f"Valor Crítico: {np.round(valor_critico, 2)}")
print(f"Intervalo de Confianza: ({np.round(limite_inferior_lifeexpectancy_i, 2)}, {np.round(limite_superior_lifeexpectancy_i, 2)})")
```

2.2

Intervalo de Confianza para el PIB
Contexto: Eres una economista y quieres calcular el intervalo de confianza al 90% para el Producto Interno Bruto (PIB) de los continentes.gdp_k

```python
media_gdp_k = df["gdp_k"].mean()

error_gdp_k = stats.sem(df["gdp_k"])

nivel_confianza_gdp_k = 0.90

grados_libertad_gdp_k = len(df["lifeexpectancy_i"]) - 1

valor_critico_gdp_k = stats.t.ppf((1 + nivel_confianza_gdp_k) / 2, df=grados_libertad_gdp_k)

limite_inferior_gdp_k = media_gdp_k - valor_critico_gdp_k * error_gdp_k
limite_superior_gdp_k = media_gdp_k + valor_critico_gdp_k * error_gdp_k

print("Intervalo de Confianza para el PIB:")
print(f"Media Muestral: {np.round(media_gdp_k, 2)}")
print(f"Error Estándar: {np.round(error_gdp_k, 2)}")
print(f"Nivel de Confianza: {nivel_confianza_gdp_k}")
print(f"Valor Crítico: {np.round(valor_critico_gdp_k, 2)}")
print(f"Intervalo de Confianza: ({np.round(limite_inferior_gdp_k, 2)}, {np.round(limite_superior_gdp_k, 2)})")
```

2.3

Intervalo de Confianza para la Tasa de Natalidad
Contexto: Trabajas en una agencia gubernamental que se ocupa de cuestiones demográficas y deseas determinar el intervalo de confianza al 99% para la tasa de natalidad promedio en Asia. 	birthrate_k

```python
df_natalidad_asia = df[df["Continent"]== "Asia"]
df_natalidad_asia
```

```python
media_birthrate_k = df_natalidad_asia["birthrate_k"].mean()

error_birthrate_k = stats.sem(df_natalidad_asia["birthrate_k"])

nivel_confianza_birthrate_k = 0.99

grados_libertad_birthrate_k = len(df_natalidad_asia["birthrate_k"]) - 1

valor_critico_birthrate_k = stats.t.ppf((1 + nivel_confianza_birthrate_k) / 2, df=grados_libertad_birthrate_k)

limite_inferior_birthrate_k = media_birthrate_k - valor_critico_birthrate_k * error_birthrate_k
limite_superior_birthrate_k = media_birthrate_k + valor_critico_birthrate_k * error_birthrate_k

print("Intervalo de Confianza para la tasa de natalidad Asia:")
print(f"Media Muestral: {np.round(media_birthrate_k, 2)}")
print(f"Error Estándar: {np.round(error_birthrate_k, 2)}")
print(f"Nivel de Confianza: {nivel_confianza_birthrate_k}")
print(f"Valor Crítico: {np.round(valor_critico_birthrate_k, 2)}")
print(f"Intervalo de Confianza: ({np.round(limite_inferior_birthrate_k, 2)}, {np.round(limite_superior_birthrate_k, 2)})")
```
